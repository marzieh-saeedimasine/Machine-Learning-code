```python
##PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

##
X = np.array([[1,2],[3,4],[5,6]])
pca=PCA(n_components=2)
X_new=pca.fit_transform(X)
print(pca.components_)
print(pca.explained_variance_)
print(pca.singular_values_)

##plot
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
fig.subplots_adjust(left=0.06, right=1, wspace=0.1)

ax[0].scatter(X[:,0],X[:,1],c='g')
ax[1].scatter(X_new[:,0],X_new[:,1],c='r')
ax[1].set( xlim=(-3, 3), ylim=(-0.01,0.01));
```

```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

##PCA on iris dataset
X=load_iris().data
Y=load_iris().target
pca=PCA(n_components=2)
X_new=pca.fit_transform(X)
print(pca.components_, pca.explained_variance_)



##plot
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
fig.subplots_adjust(left=0.06, right=1, wspace=0.1)

ax[0].scatter(X[:,0],X[:,1],c=Y)
ax[1].scatter(X_new[:,0],X_new[:,1],c=Y)
```

```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA, KernelPCA

##PCA on iris dataset
X=load_iris().data
Y=load_iris().target
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
print(pca.components_, pca.explained_variance_)


##KPCA
kpca = KernelPCA(n_components=2, kernel='linear')
X_kpca=kpca.fit_transform(X)



##plot
fig, ax = plt.subplots(1, 3, figsize=(12, 5))
fig.subplots_adjust(left=0.06, right=1, wspace=0.1)

ax[0].scatter(X[:,0],X[:,1],c=Y)
ax[1].scatter(X_pca[:,0],X_pca[:,1],c=Y)
ax[2].scatter(X_kpca[:,0],X_kpca[:,1],c=Y)
```
